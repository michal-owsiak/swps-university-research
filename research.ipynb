# **Abstract**

**Background:** Studies have consistently shown that controlled psychedelic use and mystical experiences can produce positive outcomes such as increased life satisfaction, enhanced empathy, and reduced fear of death. However, despite the extensive research, there still remain gaps in the available data. 

**Aims:** The study tested whether individuals who (H1) use psychedelics or (H2) have had a complete mystical experience exhibit (1) higher life satisfaction and (2) greater emotional empathy, along with (3) lower fear of death. It also examined if (H3) a complete mystical experience provides a better explanation for the outcomes compared to psychedelic use alone.

**Methods:** An online survey that included questions about psychedelic use and questionnaires measuring mystical experiences, fear of death, emotional empathy, and life satisfaction. A sample of 686 subjects was analyzed using two-way ANOVA and linear regression modeling. 

**Results:** Complete mystical experiences were significantly linked with lower fear of death, higher life satisfaction, and greater overall and positive-emotional empathy. Psychedelic users exhibited fear of death lower than non-users, with no significant effects on overall emotional empathy or life satisfaction. The users group also demonstrated significantly lower negative-emotional empathy than non-users.

**Conclusions:** While causal inference cannot be drawn from this study, it suggests that mystical experiences, regardless of their trigger, may potentially mitigate death anxiety, enhance life satisfaction, and foster a greater inclination to share positive emotions with others.

# **1. Brief introduction**

Psychedelic substances such as LSD, psilocybin, DMT, 5-MeO-DMT, and mescaline are known for inducing intense cognitive and emotional experiences, often described by individuals as personal breakthroughs. Similar experiences can also be triggered by other conditions, including:

- temporal lobe epilepsy, particularly in its rare ecstatic subtype;
- strokes or mechanical damage to the frontal and temporal brain lobes;
- clinical death or closely facing one's own death;
- prolonged exposure to high altitudes, especially when combined with social isolation and/or stress (e.g., during mountain climbing);
- meditation;
- religious rituals (other than ritualistic psychedelic use);
- breathing techniques like holotropic breathing;
- sensory deprivation;
- sleep deprivation;
- lucid dreaming;
- intense stress or physical exertion;
- profound experiences with works of art.

Mystical experiences are often accompanied by a profound sense of awe and are described as transcending conventional categories of time, space, logic, and personal identity.

In recent decades, research on these experiences has shifted from philosophy and religious studies to psychopharmacology, neuroscience, and other empirical disciplines. This transition has led to the development of more precise methods for studying these phenomena. The presented study's objective is to explore the associations between mystical experiences, psychedelic use, and following variables: fear of death, emotional empathy, and life satisfaction. The study aims to address recognized gaps in psychedelic and mystical experience research and to test the following hypotheses:

<br>

**H1**: Individuals using psychedelics exhibit higher levels of life satisfaction and emotional empathy and lower levels of fear of death compared to non-users.

**H2**: Individuals who have experienced a complete mystical experience show higher levels of life satisfaction and emotional empathy and lower levels of fear of death compared to those who have not.

**H3**: The combination of a complete mystical experience and psychedelic use explains a greater variance in life satisfaction, emotional empathy, and fear of death than psychedelic use alone.

<br>

These hypotheses test the idea that it is not merely the use of psychedelics that is associated with the dependent variables, but rather the mystical experiences these substances can induce. The study also examines other methods or activities that can induce mystical experiences to determine whether the experience itself, or the manner in which it is achieved, is a significant factor influencing the dependent variables.

The presented study is arguably the first to measure aforementioned variables within the context of psychedelic use and mystical experiences in a Polish research sample. The findings may contribute to the debate on the therapeutic potential of psychedelics and expand the cultural contexts of mystical experience research.


# **2. Data collection & dataset overview**

Before we proceed with data wrangling, cleaning, and analysis, it is essential to understand the dataset and the data collection process. Below is a summary of the dataset and its collection:

The online survey was created and distributed using [Qualtrics XM](https://www.qualtrics.com/). Recruitment occurred from February 17, 2023, to March 29, 2023. The survey included the following sections:

- **Homepage**

  Participants were informed about the study conditions, an overview of their participation, and an estimate of its duration. The introduction highlighted that some survey parts might cause emotional discomfort. Participants were assured of complete anonymity and informed they could withdraw at any time. Clicking the "Proceed" button at the bottom of the page indicated consent to participate.

- **Demographic Metrics**

  This section collected basic demographic data: age, gender, level of education, and spiritual worldview. Participants were also asked to choose between a masculine or feminine version of the survey. Based on their choice, subsequent content used either masculine or feminine forms of verbs and adjectives (in Polish).

- **Meditation/Mindfulness Practice**

  Participants were asked if they practice meditation or mindfulness. Those who answered "yes" were asked to estimate the number of minutes spent on this practice in the last 30 days.

- **Psychedelics Use**

  Participants selected from a list of psychedelic substances they had used at least once in their lives, with an option to add other substances. The list included: LSD (or 1P-LSD), psilocybin mushrooms (or synthetic psilocybin), ayahuasca, DMT (other than ayahuasca), 5-MeO-DMT, mescaline, ibogaine, and *Salvia divinorum*.

  Those who selected at least one substance estimated how many times they had used psychedelics and the subjective doses (microdoses, low, average, high, very high). *Salvia divinorum* (salvinorin A) and ibogaine were excluded from the psychedelics group as non-classic psychedelics.

- **Mystical Experiences**

  Participants were asked if they had experienced a state of consciousness characterized by: 
  1. An altered sense of time and/or space
  2. A sense of awe, wonder, or fear
  3. Ineffability
  4. Subjective transcendental/mystical qualities

  Negative responses bypassed subsequent items and the Mystical Experience Questionnaire (MEQ30), moving directly to dependent variable questionnaires.

  Affirmative responses led to a multiple-choice list to specify the circumstances of the experience. If a psychedelic substance was chosen, participants specified which compound triggered the most intense experience. For multiple circumstances, they indicated the first and most intense experience. The last item asked how long ago the most intense experience occurred.

- **Revised Mystical Experience Questionnaire (MEQ30)**

  This 30-item questionnaire measures the intensity of mystical experiences across four dimensions: *Mystical*, *Positive Affect*, *Transcendence of Time and Space*, and *Ineffability*. Participants rated their experience on a 6-point Likert scale. 

  The Polish translation (α = 0.95, N = 515) was used. Confirmatory factor analysis showed a structural difference from the original. The item "Experience of amazement" loaded onto the *Ineffability* dimension instead of *Positive Affect*. Retaining the original structure preserved slightly better reliability: (α<sub>mysticism</sub> = 0.95; α<sub>positive affect</sub> = 0.80; α<sub>transcendence</sub> = 0.85; α<sub>ineffability</sub> = 0.79).

- **Perth Empathy Scale (PES)**

  This tool measures cognitive and affective empathy, categorized into positive and negative emotions, with four subscales. The study used the *Positive* and *Negative Affective Empathy* subscales (α<sub>positive-affective</sub> = 0.70; α<sub>negative-affective</sub> = 0.72, N = 676), contributing to *Overall Affective Empathy* (α = 0.73; 10 items). Each item used a 5-point Likert scale. The Polish version by Paweł Larionow and Karolina Mudło-Głagolska was utilized.

- **Satisfaction with Life Scale (SWLS)**

  This scale includes 5 items rated on a 7-point Likert scale, measuring life satisfaction (example item: *If I could live my life over, I would change almost nothing*). The Polish translation by Konrad Jankowski was used (α = 0.86, N = 676).

- **Death Attitude Profile - Revised (DAP-R-PL)**

  This tool measures attitudes toward death, including types of death acceptance and avoidance, and fear of death. Only the *Fear of Death* subscale (example item: *Death is undoubtedly an unpleasant experience*) was used in the Polish adaptation by Paweł Brudek and colleagues. This subscale contains 10 items rated on a 7-point Likert scale (α = 0.92, N = 676).

- **Subjectively Perceived Influence**

  This section was for participants who reported a mystical experience. They answered three questions about the impact of the experience on empathy, life satisfaction, and fear of death, rated on a 5-point Likert scale. Participants also had the option to share their experience in an additional text field.

# **4. Data wrangling and cleaning**

## **4.1. Joining female and male variables into one dataframe**

Regarding the incomplete records mentioned earlier, we will set them aside for now and proceed to aggregate the `_F` and `_M` variables into a consolidated dataframe.

First, we will convert all non-numeric records to zeros (we can apply it for all the columns - string variables included - as we don't need descriptive records for our analysis). Having numeric values, we will sum the corresponding `_F` and `_M` values together, storing the combined columns in a separate dataframe. This dataframe will then be concatenated with the original one. Finally, we will drop all `_F` and `_M` columns, which will leave us with the desired dataset for further wrangling.

```python
columns_to_process = [col[:-2] for col in df.columns if col.endswith("_M")]
combined_columns = {}

for col in columns_to_process:
    male_col = f"{col}_M"
    female_col = f"{col}_F"
    
    male_variables = pd.to_numeric(df[male_col], errors="coerce").fillna(0)
    female_variables = pd.to_numeric(df[female_col], errors="coerce").fillna(0)
    
    combined_columns[col] = male_variables + female_variables

combined_df = pd.DataFrame(combined_columns)

df = pd.concat([df, combined_df], axis=1)

df.drop(columns=[f"{col}_M" for col in columns_to_process] + [f"{col}_F" for col in columns_to_process], inplace=True)

df.head(10)
```

Now we can see that all the NaN values have been replaced with zeros and that we no longer have duplicated set of female and male variables.

## **4.2. Recoding the variables**

The next crucial step involves appropriately recoding the scores from the questionnaires. While none of them include reversed items, their scales' coding needs some adjustments (apart from PES questionnaire).

**MEQ30**

When exporting data from Qualtrics (the survey creation tool) to an *.xlsx* file, Likert-type responses are automatically coded as follows: the first answer on the scale is marked as `1`, the second as `2`, and so on. However, for the MEQ30 questionnaire, the original scale ranges from 0 to 5. Therefore, we need to recode the current values accordingly. We'll identify all columns starting with `MEQ30`, create a dictionary where each key corresponds to a value that needs replacement, and then apply this dictionary to replace the values in these columns.

```python
meq30_columns = [col for col in df.columns if col.startswith("MEQ30")]

replacement_mapping = {
    1: 0, 
    2: 1, 
    3: 2, 
    4: 3, 
    5: 4, 
    6: 5
}

df[meq30_columns] = df[meq30_columns].replace(replacement_mapping)
```

**SWLS and DAP-R**

For the SWLS and DAP-R questionnaires, our goal is to obtain scales starting from 1, so there shouldn't be an issue as with the MEQ30. However, these questionnaires are structured such that the first answer on the list reflects **total agreement** with the statement of each item (e.g., *I am satisfied with my life*). Therefore, to ensure that a higher level of agreement corresponds to a higher score on the scale, we need to reverse the scoring of all variables in these two questionnaires. We will employ the same approach used for the MEQ30 described above.

```python
swls_columns = [col for col in df.columns if col.startswith("SWLS")]
dap_r_columns = [col for col in df.columns if col.startswith("DAP_R")]

reverse_mapping = {
    1: 7, 
    2: 6, 
    3: 5,
    5: 3, 
    6: 2, 
    7: 1
}

df[swls_columns] = df[swls_columns].replace(reverse_mapping)
df[dap_r_columns] = df[dap_r_columns].replace(reverse_mapping)
```

## **4.3. Calculating the summed results**

Now that we have all the scales ready, we can proceed to summing the values up. Let's start with the MEQ30.

### **4.3.1. MEQ30**

First, we need to create a dictionary based on the questionnaire's scoring instructions. For each subscale (`key`), we will provide a list of items belonging to that subscale. Next, we will create a `for` loop to iterate through all the items in each subscale and sum them. This loop will create new columns with the summed values for each subscale.

```python
MEQ30_subscales = {
    "Mystical": [4, 5, 6, 9, 14, 15, 16, 18, 20, 21, 23, 24, 25, 26, 28],
    "Positive Mood": [2, 8, 12, 17, 27, 30],
    "Transcendence": [1, 7, 11, 13, 19, 22],
    "Ineffability": [3, 10, 29]
}

for subscale, items in MEQ30_subscales.items():
    df.loc[:, subscale] = df[[f"MEQ30_{item}" for item in items]].sum(axis=1)
```

Let's make sure that now the last 4 columns in our dataframe are actually the MEQ30 subscales' scorings.

```python
df.iloc[:10, -4:]
```

As seen above, this approach seems to be working. However, we are still not finished with the MEQ30. The crucial information we are looking for is whether a person has undergone a *complete mystical experience*. According to the authors of the questionnaire, this is defined by scoring at least 60% of the points in each of the four subscales.

For that matter, we will create a dictionary to assign the maximum scoring value to each subscale. This can be done by taking the lengths of the lists from our previous `MEQ30_subscales` dictionary and multiplying them by 5 (since our scale for this questionnaire ranges from 0 to 5). Next, we will take the scores displayed in the table above and divide them by the calculated maximum values, checking if all the four results are at least 60%. Finally, we will convert resulting Boolean flags into numeric values.

```python
max_score = {
    "Mystical": len(MEQ30_subscales["Mystical"]) * 5,
    "Positive Mood": len(MEQ30_subscales["Positive Mood"]) * 5,
    "Transcendence": len(MEQ30_subscales["Transcendence"]) * 5,
    "Ineffability": len(MEQ30_subscales["Ineffability"]) * 5
}

df["MEQ_complete"] = (
    (df["Mystical"] / max_score["Mystical"] >= 0.6) &
    (df["Positive Mood"] / max_score["Positive Mood"] >= 0.6) &
    (df["Transcendence"] / max_score["Transcendence"] >= 0.6) &
    (df["Ineffability"] / max_score["Ineffability"] >= 0.6)
).astype(int)

df.iloc[:10, -1:]
```

### **4.3.2. PES**

To sum up the overall emotional empathy score, we will need to iterate through all 10 PES items, excluding item number 9, as it serves as an additional question to anchor the participant's attention.

```python
pes_items = [f"PES_{i}" for i in range(1, 12) if i != 9]

df["PES_sum"] = df[pes_items].sum(axis=1)
```

Additionally, it will be useful to have separate summed scores for positive and negative emotions. Therefore, we need to implement the same approach as we did with the MEQ30.

```python
PES_subscales = {
    "PES_positive": [2, 4, 6, 8, 11],
    "PES_negative": [1, 3, 5, 7, 10]
}

for subscale, items in PES_subscales.items():
    df.loc[:, subscale] = df[[f"PES_{item}" for item in items]].sum(axis=1)
```

Now we have three summed columns for the PES questionnaire:

```python
df.iloc[:10, -3:]
```

### **4.3.3. SWLS and DAP-R**

Similarly to the PES, we will sum the items of the DAP-R, and then of the SWLS questionnaire. For the DAP-R, we will exclude the 4th item as it is a control question.

```python
dap_r_items = [f"DAP_R_{i}" for i in range(1, 9) if i != 4]
swls_items = [f"SWLS_{i}" for i in range(1, 6)]

df["DAP_R_sum"] = df[dap_r_items].sum(axis=1)
df["SWLS_sum"] = df[swls_items].sum(axis=1)

df.iloc[:10, -2:]
```

### **4.4.2. Dropping records with unconscious responses**

Within the original questionnaire items, there were 2 control questions mixed in. The purpose of these questions was to identify participants who may have been marking random answers.

For PES, the control question is:
- "If you are reading this, please mark ***Often***", which corresponds to the expected value of `4`.

For DAP-R, the control question is:
- "If you are reading this, please mark ***Rather disagree***", which corresponds to the expected value of `3` on our recoded scale.

### **4.4.3. Dropping the outliers**

During the research design phase, it was determined to exclude outliers that fall outside `2.5` standard deviations (*SD*) from the mean scores of dependent variables. This approach aims to enhance the likelihood of obtaining normally distributed results.

To begin, we need to calculate the *Z*-scores for each record. The *Z*-score indicates how many *SD* a particular data point deviates from the mean score. Here is the formula for calculating it:

<p style="font-size: 22px; font-weight: bold">
    <i>Z = (X - μ) ÷ σ</i>
</p>

where:

***X*** - the individual data point,

***μ*** - the mean score of the data,

***σ*** - the *SD* value.

We will apply this formula to our dependent variables and create new columns to store these calculated *Z*-scores. Since the *Z* only reflects the magnitude of deviation from the mean, without indicating direction (above or below), we will use the NumPy function `abs` to obtain absolute (positive) values of the scores.

```python
df.loc[:, "PES_Z-score"] = np.abs((df["PES_sum"] - df["PES_sum"].mean()) / df["PES_sum"].std())
df.loc[:, "SWLS_Z-score"] = np.abs((df["SWLS_sum"] - df["SWLS_sum"].mean()) / df["SWLS_sum"].std())
df.loc[:, "DAP_R_Z-score"] = np.abs((df["DAP_R_sum"] - df["DAP_R_sum"].mean()) / df["DAP_R_sum"].std())

df.iloc[:10, -3:]
```

As demonstrated in the sample above, certain scores exceed `2.5` standard deviations from the mean (e.g., row 9 for PES and row 5 for DAP-R). To maintain consistency with our established criteria, these outliers will be removed from the dataset.

```python
df_outliers = df[(df["PES_Z-score"] > 2.5) | (df["SWLS_Z-score"] > 2.5) | (df["DAP_R_Z-score"] > 2.5)]

df = df[~df.index.isin(df_outliers.index)]
df.reset_index(inplace=True)

print(f"We have filtered out {df_outliers.shape[0]} outliers.")
```

# **5. Data analysis**

First, we will define our four research groups based on binary division of two independent variables (*psychedelics use* and *complete mystical experience*):

  - **Group 1:** 

    Psychedelics use - **Yes** | Complete mystical experience - **No**
  
  - **Group 2:** 

    Psychedelics use - **No** | Complete mystical experience - **No**

  - **Group 3:** 

    Psychedelics use - **Yes** | Complete mystical experience - **Yes**
  
  - **Group 4:** 

    Psychedelics use - **No** | Complete mystical experience - **Yes**

To facilitate further analyses, we'll create a function that takes a dataframe and a specified dependent variable as arguments, returning the data divided into desired groups. The list of dependent variables is also provided below.

```python
def define_groups(df, dependent_variable):
    group_1 = df[(df["compound_never"] == 0) & (df["MEQ_complete"] == 0)][dependent_variable]
    group_2 = df[(df["compound_never"] == 1) & (df["MEQ_complete"] == 0)][dependent_variable]
    group_3 = df[(df["compound_never"] == 0) & (df["MEQ_complete"] == 1)][dependent_variable]
    group_4 = df[(df["compound_never"] == 1) & (df["MEQ_complete"] == 1)][dependent_variable]

    return group_1, group_2, group_3, group_4


dependent_variables = ["PES_sum", "SWLS_sum", "DAP_R_sum"]
```

## **5.1. Testing parametric assumptions**

All the parametric tests, such as ANOVA, rely on specific assumptions that must be satisfied for the results to be valid. These assumptions typically include the normality of data distributions within groups, homogeneity of variances, and independence of observations. In this section, we will systematically evaluate these assumptions to ensure the robustness and reliability of subsequent statistical analyses.

### **5.1.1. Tests for homogeneity of variance**

One of the common assumptions of parametric tests like ANOVA is that all groups exhibit statistically equal variance values. To assess this assumption in our dataset, we will conduct a series of three Levene's tests using the SciPy `levene` function. These tests will utilize our `define_groups` function and the `dependent_variables` list.

```python
def perform_levenes_test(group_1, group_2, group_3, group_4):
    stat, p_value = levene(
        group_1,
        group_2,
        group_3,
        group_4,
        center="mean"
    )
    
    return stat, p_value

results = {}

for variable in dependent_variables:
    group_1, group_2, group_3, group_4 = define_groups(df, variable)
    stat, p_value = perform_levenes_test(group_1, group_2, group_3, group_4)
    results[variable] = (stat, p_value)

for variable, (stat, p_value) in results.items():
    print(f"Levene's test for {variable}: statistic: {stat:.3f}; p-value: {p_value:.3f}")
```

We observe that the assumptions for emotional empathy (PES) and satisfaction with life (SWLS) are met. However, significant variance differences are evident in the fear of death variable, with a *p*-value well below 0.05. But does this discrepancy definitively preclude the use of analysis of variance? Let's explore further.

The *F*-test (ANOVA statistic) demonstrates high robustness to variance heterogeneity. A critical indicator of this robustness is the *variance ratio* (VR): the ratio of the largest to smallest variance value among the compared groups. If our VR value is 1.5 or less, and we are comparing at least 3 groups, it indicates **100%** robustness against significant Levene's test results.

Now, let's calculate the VR value for the fear of death variance:

```python
def calculate_VR(group_1, group_2, group_3, group_4):
    variances = [group.var() for group in [group_1, group_2, group_3, group_4]]
    max_var = max(variances)
    min_var = min(variances)
    VR = max_var / min_var
    
    return VR

group_1, group_2, group_3, group_4 = define_groups(df, variable)

DAP_R_VR = calculate_VR(group_1, group_2, group_3, group_4)

print(f"The variance ratio (VR) value for fear of death variable is {DAP_R_VR:.2f}.")
```

The VR value is less than 1.5. Good! We can now proceed to test the normality of data distribution in our groups.

### **5.1.2. Tests for normal distribution**

We will perform a series of Shapiro-Wilk tests to assess whether the distribution of each dependent variable within each group deviates significantly from a normal distribution. For this purpose, we will utilize the SciPy `shapiro` function.

```python
for var in dependent_variables:
    group_1, group_2, group_3, group_4 = define_groups(df, var)
    
    stat_1, p_value_1 = shapiro(group_1)
    stat_2, p_value_2 = shapiro(group_2)
    stat_3, p_value_3 = shapiro(group_3)
    stat_4, p_value_4 = shapiro(group_4)

    print(f"Shapiro-Wilk test for {var}:\n")
    print(f"Group 1 - statistic: {stat_1:.3f}, p-value: {p_value_1:.3f}")
    print(f"Group 2 - statistic: {stat_2:.3f}, p-value: {p_value_2:.3f}")
    print(f"Group 3 - statistic: {stat_3:.3f}, p-value: {p_value_3:.3f}")
    print(f"Group 4 - statistic: {stat_4:.3f}, p-value: {p_value_4:.3f}\n")
```

We can see that the assumption of normal distribution is met only in half of the cases (all the PES groups as well as SWLS groups 2 and 4). However, ANOVA is relatively robust to violations of normality, especially with larger sample sizes due to the Central Limit Theorem. If the samples in each grup are of sufficient size (typically *N* > 30), the ANOVA can still produce reliable results even if the data are not perfectly normally distributed.

Therefore, considering our sample sizes...

```python
group_1_size = len(df[(df["compound_never"] == 0) & (df["MEQ_complete"] == 0)])
group_2_size = len(df[(df["compound_never"] == 1) & (df["MEQ_complete"] == 0)])
group_3_size = len(df[(df["compound_never"] == 0) & (df["MEQ_complete"] == 1)])
group_4_size = len(df[(df["compound_never"] == 1) & (df["MEQ_complete"] == 1)])

print("Research sample sizes:\n")
print(f"Group 1 - {group_1_size}")
print(f"Group 2 - {group_2_size}")
print(f"Group 3 - {group_3_size}")
print(f"Group 4 - {group_4_size}")
```

...we can trust the results that we will obtain from the analysis of variance. There is a visible imbalance in the 4th group size, but we will address this issue below.

## **5.2. Series of two-way ANOVA**

The main analysis in our research will be a **two-way analysis of variance** (ANOVA based on two factors). Conducting a two-way ANOVA for each dependent variable will allow us to determine if there are significant main effects of **(1) psychedelic use** and **(2) complete mystical experience**, as well as whether there is a significant interaction between these two factors. Testing the main effects will inform us whether to reject or accept the hypotheses **H1** and **H2**.

We will use the ANOVA model from the `pingouin` (`pg`) package. The `pg.anova` function uses *Type III sums of squares*, which is the preferred method for factorial (e.g., two-way) ANOVA, especially when the data is unbalanced. It is also the default type used in *IBM SPSS Statistics* software, so this way we will maintain consistency with the originally used method.

As noted, our sample sizes are unequal, with the 4th group being significantly smaller than the other groups. Type III is considered robust to such imbalances.

But before that, let's print the mean values of our variable for each group condition. This will help us understand the direction of the assumed effects. Here, we will also create a `lambda` function to output formatted results.

```python
psychedelic_users = df[df["compound_never"] == 0]
psychedelic_non_users = df[df["compound_never"] == 1]

mean_psychedelic_users = psychedelic_users["DAP_R_sum"].mean()
mean_psychedelic_non_users = psychedelic_non_users["DAP_R_sum"].mean()

complete_mystical = df[df["MEQ_complete"] == 1]
no_complete_mystical = df[df["MEQ_complete"] == 0]

mean_complete_mystical = complete_mystical["DAP_R_sum"].mean()
mean_no_complete_mystical = no_complete_mystical["DAP_R_sum"].mean()

print("Fear of death mean scores for psychedelic use condition:\n")
print(f"Psychedelic non-users: {mean_psychedelic_non_users:.2f}")
print(f"Psychedelic users: {mean_psychedelic_users:.2f}\n")
print("Fear of death mean scores for complete mystical experience condition:\n")
print(f"No complete mystical experience: {mean_no_complete_mystical:.2f}")
print(f"Complete mystical experience: {mean_complete_mystical:.2f}")
```

## **5.4. Linear regression modeling**

To test our H3 hypothesis, we will use hierarchical regression modeling. In Model 1, we will use the binary variable for psychedelic use/non-use as a predictor. In Model 2, we will add the binary variable for complete mystical experience/no such experience to the hierarchy and calculate the *p*-values for *ΔR<sup>2</sup>* (the change in *R<sup>2</sup>* between Models 1 and 2).

### **5.4.1. Fear of death**

```python
model1 = ols("DAP_R_sum ~ compound_never", data=df).fit()
model2 = ols("DAP_R_sum ~ compound_never + MEQ_complete", data=df).fit()

delta_r2 = model2.rsquared - model1.rsquared
delta_df = model2.df_model - model1.df_model
resid_df2 = model2.df_resid
delta_f = (delta_r2 / delta_df) / ((1 - model2.rsquared) / resid_df2)
p_value = f.sf(delta_f, delta_df, model2.df_resid)
formatted_p_value = "< 0.001" if p_value < 0.001 else f"{p_value:.3f}"

summary_table = pd.DataFrame({
    'R squared for Model 1': [f"{model1.rsquared:.3f}"],
    'R squared for Model 2': [f"{model2.rsquared:.3f}"],
    'ΔR squared': [f"{delta_r2:.3f}"],
    'F for change': [f"{delta_f:.3f}"],
    'p-value for change': [formatted_p_value]
})

summary_table
```

### **5.4.2. Emotional empathy**

```python
model1 = ols("PES_sum ~ compound_never", data=df).fit()
model2 = ols("PES_sum ~ compound_never + MEQ_complete", data=df).fit()

delta_r2 = model2.rsquared - model1.rsquared
delta_df = model2.df_model - model1.df_model
resid_df2 = model2.df_resid
delta_f = (delta_r2 / delta_df) / ((1 - model2.rsquared) / resid_df2)
p_value = f.sf(delta_f, delta_df, model2.df_resid)
formatted_p_value = "< 0.001" if p_value < 0.001 else f"{p_value:.3f}"

summary_table = pd.DataFrame({
    'R squared for Model 1': [f"{model1.rsquared:.3f}"],
    'R squared for Model 2': [f"{model2.rsquared:.3f}"],
    'ΔR squared': [f"{delta_r2:.3f}"],
    'F for change': [f"{delta_f:.3f}"],
    'p-value for change': [formatted_p_value]
})

summary_table
```

### **5.4.2. Life satisfaction**

```python
model1 = ols("SWLS_sum ~ compound_never", data=df).fit()
model2 = ols("SWLS_sum ~ compound_never + MEQ_complete", data=df).fit()

delta_r2 = model2.rsquared - model1.rsquared
delta_df = model2.df_model - model1.df_model
resid_df2 = model2.df_resid
delta_f = (delta_r2 / delta_df) / ((1 - model2.rsquared) / resid_df2)
p_value = f.sf(delta_f, delta_df, model2.df_resid)
formatted_p_value = "< 0.001" if p_value < 0.001 else f"{p_value:.3f}"

summary_table = pd.DataFrame({
    'R squared for Model 1': [f"{model1.rsquared:.3f}"],
    'R squared for Model 2': [f"{model2.rsquared:.3f}"],
    'ΔR squared': [f"{delta_r2:.3f}"],
    'F for change': [f"{delta_f:.3f}"],
    'p-value for change': [formatted_p_value]
})

summary_table
```

Upon analysis, the *p*-value for the change in predictive power of the model was found to be statistically significant for each dependent variable, thus validating the **H3** hypothesis. However, it is noteworthy that the *R<sup>2</sup>* values remained quite low for emotional empathy (1%) and life satisfaction (2%). In contrast, Model 2 was able to explain 11% of the observed variance in fear of death.

# **7. Conclusions**

**Mystical experiences are related to fear of death, empathy, and life satisfaction levels**

People with a complete mystical experience are characterized by significantly lower fear of death, significantly higher emotional empathy, and significantly higher life satisfaction compared to others. This constitutes the first main finding of this study, fully confirming hypothesis H2. Among all three variables, the effect on fear of death was the most pronounced. The difference in average scores between people who have never taken psychedelics and never had a complete mystical experience, and those who have taken psychedelics and had such an experience (26.61 points vs. 17.68 points), amounted to 8.93 points on a 43-point scale (21% of the scale). In the case of life satisfaction, the difference between the extreme groups with and without a full mystical experience was 3.24 points on a 31-point scale (11% of the scale), and in the case of empathy, it was 2.18 points on a 41-point scale (5% of the scale). The condition of a complete mystical experience also significantly increased the prediction for each of the three dependent variables compared to the condition of psychedelic use alone, which constitutes the second main finding of the study, fully confirming hypothesis H3.

**Psychedelic use is related to fear of death levels**

The last main conclusion of the study is as follows: people who have consumed a classic psychedelic substance at least once in their life have significantly lower fear of death compared to others. This finding partially confirms hypothesis H1, which was not confirmed for the variables (1) overall emotional empathy and (2) life satisfaction. For these variables, no significant differences were obtained, but they tended to approach confirmation of the opposite hypotheses not established in the study – people who have taken a psychedelic substance at least once tended to have lower scores on the scales of life satisfaction and emotional empathy compared to others. However, these tendencies were not present in the case of empathy towards positive emotions.

**Summary**

According to the study results, experiencing a complete mystical experience is associated with lower fear of death, higher life satisfaction, and a greater tendency to empathize with positive emotions. Mystical experiences significantly increased the predictive power of the studied variables compared to psychedelic use alone. It was observed that the use of these substances differentiated individuals in terms of fear of death, but only if they had not experienced a complete mystical experience. Those who had such an episode exhibited consistently low levels of fear, regardless of psychedelic use. Users of these substances showed a lower tendency to empathize with negative emotions.

The connections with empathy and life satisfaction are independent and differ between psychedelic use and complete mystical experiences. The separation of these associations, as well as indicating an interaction effect in the case of fear of death, could be an indirect but valuable contribution to the debate on the therapeutic potential of substances that induce mystical experiences. Although available experimental studies correspond well with the results presented here, these results do not allow for the confirmation of causal relationships. To establish these, longitudinal studies with an appropriate methodological design are needed in the future.
